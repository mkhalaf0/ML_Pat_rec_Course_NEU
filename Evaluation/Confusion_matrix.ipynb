```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

# 1 = spam (positive)
# 0 = legitimate (negative)

# TP = 30: actually spam, predicted spam
# FN = 10: actually spam, predicted legit
# FP = 5: actually legit, predicted spam
# TN = 55: actually legit, predicted legit

y_true =  [1]*40 + [0]*60
y_pred =  [1]*30 + [0]*10 + [1]*5 + [0]*55

# Manual counts
TP = 0
TN = 0
FP = 0
FN = 0

for true, pred in zip(y_true, y_pred):
    if true == 1 and pred == 1:
        TP += 1
    elif true == 0 and pred == 0:
        TN += 1
    elif true == 0 and pred == 1:
        FP += 1
    elif true == 1 and pred == 0:
        FN += 1

print("Step 1: Confusion Matrix Values")
print("TP =", TP)
print("TN =", TN)
print("FP =", FP)
print("FN =", FN)

print("\n Check")
print("Total =", TP + TN + FP + FN)
print("Actual spam =", TP + FN)
print("Actual legit =", FP + TN)
print("Predicted spam =", TP + FP)
print("Predicted legit =", FN + TN)

# Step 2: Accuracy
accuracy = (TP + TN) / (TP + TN + FP + FN)
print("\nStep 2: Accuracy")
print("Accuracy = (TP + TN) / Total")
print(f"Accuracy = ({TP} + {TN}) / 100 = {accuracy:.2f}")

# Step 3: Precision
precision = TP / (TP + FP)
print("\nStep 3: Precision")
print("Precision = TP / (TP + FP)")
print(f"Precision = {TP} / ({TP} + {FP}) = {precision:.3f}")

# Step 4: Recall
recall = TP / (TP + FN)
print("\nStep 4: Recall")
print("Recall = TP / (TP + FN)")
print(f"Recall = {TP} / ({TP} + {FN}) = {recall:.2f}")

```

```python
# Step 5: Using sklearn
print("\nStep 5: Using sklearn")

cm = confusion_matrix(y_true, y_pred, labels=[1, 0])
print("Confusion Matrix:")
print(cm)

print("\nRows = Actual, Columns = Predicted")
print("            Predicted Spam   Predicted Legit")
print("Actual Spam      ", cm[0][0], "              ", cm[0][1])
print("Actual Legit     ", cm[1][0], "               ", cm[1][1])

print("\nSklearn Metrics")
print("Accuracy =", accuracy_score(y_true, y_pred))
print("Precision =", precision_score(y_true, y_pred))
print("Recall =", recall_score(y_true, y_pred))
```
